```python
# Using the ExcelFile class to read multiple sheets (Excel file is only passed once)

# Better performance than read_excel (there, the Excel file is passed every time, hence the file is loaded again for every sheet)

xlsx = pd.ExcelFile("prueba.xlsx")
parts_sheets = []
for sheet in xlsx.sheet_names:
    parts_sheets.append(xlsx.parse(sheet))
all_df = pd.concat(parts_sheets)

all_df.reset_index(drop=True, inplace=True) # reset the values of 

len(all_df) # 3 files = 50 + 75 + 20 = 145
print(all_df)
```

```python
# ---------------------------------------------------------------------------------------
# --------------------------- TEST = NO NEED TO RUN IT ----------------------------------
# ---------------------------------------------------------------------------------------


# I'M IMPORTING A FILE FROM JONATHAN (TRANSPORTATION)
# BICYCLE RENTAL IN TAIPEI IN 2017

# The thing is that numeric values have the symbol \t in the middle. We need to remove it.
# Check the uncorrected data:

import pandas as pd
df_transportation = pd.read_csv("Bicycle Facilities and Rental Situation in Taipei City 2017 by  month.csv")
print(df_transportation)
```

```python
# ---------------------------------------------------------------------------------------
# --------------------------- TEST = NO NEED TO RUN IT ----------------------------------
# ---------------------------------------------------------------------------------------

# So I'm rewriting the information in a new CSV file that I'm creating in the same folder,
# with the \t removed.

# CSV FORMAT 
import csv
import string
import re

input_file = open('Bicycle Facilities and Rental Situation in Taipei City 2017 by  month.csv', 'r')
output_file = open('fixformat.csv', 'w')
data = csv.reader(input_file)
writer = csv.writer(output_file,quoting=csv.QUOTE_ALL)# dialect='excel')
specials = '\t'

for line in data:
    line = [value.replace(specials, '') for value in line]
    writer.writerow(line)


input_file.close()
output_file.close()

# open the new created file with corrected numeric values:

df_transportation = pd.read_csv("fixformat.csv")
print(df_transportation)
```

```python
# ---------------------------------------------------------------------------------------
# --------------------------- TEST = NO NEED TO RUN IT ----------------------------------
# ---------------------------------------------------------------------------------------

# SAME BUT REMOVING NAMES, ONLY LEAVING ALFANUMERIC CHARACTER,
# SINCE WE KNOW EACH INDEX CORRESPONDS TO A DIFFERENT MONTH

input_file = open('Bicycle Facilities and Rental Situation in Taipei City 2017 by  month.csv', 'r')
output_file = open('fixformat.csv', 'w')
data = csv.reader(input_file)
writer = csv.writer(output_file,quoting=csv.QUOTE_ALL)# dialect='excel')
specials = '\t'

for line in data:
    line = [value.replace(specials, '') for value in line]
    line = [re.sub("[^0-9]", "", str(line)) for value in line]
    writer.writerow(line)


input_file.close()
output_file.close()

# open the new created file with corrected numeric values:

df_transportation = pd.read_csv("fixformat.csv")
print(df_transportation)
```
